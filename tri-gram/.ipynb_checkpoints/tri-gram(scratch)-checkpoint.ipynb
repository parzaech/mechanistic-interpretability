```python
import numpy as np
import pandas as pd
import matplotlib as pl
from collections import defaultdict
```

```python
with open('shaky.txt', 'r', encoding='utf-8') as f:
    text = f.read() 
```

```python
print(len(text))
```

```python
#tokenize the corpus and add [START] and [STOP] at every begining and ending of sentences
```

```python
print(text[:1000])
```

```python
# we are removing speaker labels
```

```python
lines = text.split("\n")
```

```python
dialouge = [line.strip() for line in lines if ':' not in line and line.strip()]
```

```python
print([dialouge[i] for i in range(10)])
```

```python
import re

def tokigen(text):
    return re.findall(r"\w+|\S" , text.lower())
```

```python
tokens = []
for line in dialouge: 
    
    linetoken = tokigen(line)
    add = ["<start>"] + linetoken + ["<end>"]
    tokens.extend(add)
```

```python
print([tokens[i] for i in range(100)])
```

```python
tokens = [token for token in tokens if token.isalnum() or token in {'<start>','<end>'}]
```

```python
print([tokens[i] for i in range(100)])
```

```python
vocab = set(tokens)
```

```python
print(vocab)
```

```python
v = len(vocab)
```

```python
trigram_counts = defaultdict(int)
bigram_counts = defaultdict(int)

for i in range(len(tokens) - 2):
    trigram = (tokens[i],tokens[i+1],tokens[i+2])
    bigram = (tokens[i],tokens[i+1])
    
    trigram_counts[trigram] += 1
    bigram_counts[bigram] += 1
```

```python
from itertools import islice


for trigram, count in islice(trigram_counts.items(), 100):
    print(trigram)
```

```python
# Filter trigrams with count 1 and count them
count_of_ones = len([trigram for trigram, count in trigram_counts.items() if count == 1])

print("Number of trigrams with count 1:", count_of_ones)
```

```python
# Find the trigram with the maximum count
max_trigram = max(trigram_counts.items(), key=lambda x: x[1])

print("Trigram with maximum count:", max_trigram)
```

```python
#trigram probability => P([w2|w1,w0]) = C(w2,w1,w0)/C(w1,w0)
```

```python
trigram_prob = {}
for trigram, count in trigram_counts.items():
    bigram = (trigram[0],trigram[1])
    if bigram in bigram_counts:
        trigram_prob[trigram] = (count+1)/(bigram_counts[bigram]+v)
    else:
        trigram_prob[trigram] = 1/v
```

```python
# text generation with the highest trigram probability
```

```python
def shaky_predict(word1,word2):
    candidates = [trigram for trigram in trigram_prob if trigram[0] == word1 and trigram[1] == word2]
    if not candidates:
        return none
    else:
        best_trigram = max(candidates, key=lambda trigram: trigram_prob[trigram])
        return best_trigram[2]
```
